# **6. Data Pre-Processing**
---
 - Here we will perform **Data Preprocessing** on our dataset, to make the data usable for **EDA**.

**Step1 :** Minimum and maximum ages are 13 & 113 respectively, can we come up with an range to fill the zeros of other columns. Eg: “age_group” as 13-17, 18-24,  25-35, 36-50, 51-80, Above 80

```python
data['age_group'] = data['age']
data['age_group'][ data.age < 18 ] = '13 to 17'
data['age_group'][ (data.age < 25) & (data.age >17) ] = '18 to 24'
data['age_group'][ (data.age < 36) & (data.age >24) ] = '25 to 35'
data['age_group'][ (data.age < 51) & (data.age >35) ] = '36 to 50'
data['age_group'][ (data.age < 81) & (data.age >50) ] = '51 to 80'
data['age_group'][ data.age >80 ] = 'Above 80'
data['age_group'].unique()
data.tail(10)
```

**Step2 :**  Missing Value of gender can be replaced with "Others" and missiing values tenure can be replaced with median for total tenure


```python
data['gender'].fillna('others',inplace=True)
data.gender.unique()
```

**Step3 :** “like” & “mobile_likes” can be combined to “total_likes” and can be used to analyze the overall likes based on age group, -  – Multivariate- Gender,Age group & Count

```python
data['total_likes'] = data['likes'] + data['mobile_likes']
data.head(5)
```

**Step4 :**  likes_received (γ1 = 112.0745682), mobile_likes_received (γ1 = 107.5312999) & www_likes_received (γ1 = 126.257317) are highly Skewed, we can combine them together as “total_likes_received” and the same can be used to analyze the overall likes received based on Age, Gender & Count

```python
data['total_likes_received'] = data['likes_received'] + data['mobile_likes_received'] + data['www_likes_received']
data.tail(5)
```

**Step5 :** Finding the records that can be dropped

```python
data.loc[(data['friend_count']==0) & (data['total_likes']==0) & (data['total_likes_received']==0)]
```

The records which have "friend_count", "total_likes" and "total_likes_received" can be dropped since they do not contribute to the analysis

**Step6 :** Droping the records found in Step #5

```python
newdata_df = data.copy()
```

```python
newdata_df.drop(newdata_df[(newdata_df['friend_count'] == 0) & (newdata_df['total_likes'] == 0)
                           & (newdata_df['total_likes_received'] == 0)].index, inplace = True)
```

**The records with 0 values are dropped and a new data frame newdata_df is ready for further preprocessing**

**Final Data for Post-Profiling is ready!**
